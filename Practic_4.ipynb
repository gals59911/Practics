```python
from zeep import Client

def conversion_wsdl(url):
  wsdl=url+"?wsdl"
  return wsdl

def Numbers():
  url="https://www.dataaccess.com/webservicesserver/NumberConversion.wso"
  client=Client(wsdl=conversion_wsdl(url))
  print("Выберите метод: \n","1) Число в слово \n", "2) Число долларов")
  k=int(input())
  match k:
      case 1:
        print(client.service.NumberToWords(input()))
      case 2:
        print(client.service.NumberToDollars(input()))
      case _:
        print("Error")

def Calculator():
  url="http://www.dneonline.com/calculator.asmx"
  client=Client(wsdl=conversion_wsdl(url))
  print("Выберите метод: \n","1) Сумма \n", "2) Делениe \n", "3) Умножение \n", "4) Разность ")
  k=int(input())
  match k:
      case 1:
        print(client.service.Add(input(),input()))
      case 2:
        print(client.service.Divide(input(),input()))
      case 3:
        print(client.service.Multiply(input(),input()))
      case 4:
        print(client.service.Subtract(input(),input()))
      case _:
        print("Error")

def Continents():
  url="http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso"
  client=Client(wsdl=conversion_wsdl(url))
  print(client.service.ListOfContinentsByName())

def Countries():
   url="http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso"
   client=Client(wsdl=conversion_wsdl(url))
   print("Выберите метод: \n","1) Список стран \n", "2) Столица страны \n", "3) Валюта страны \n", "4) Флаг \n","5) Международный телефонный код для копии страны")
   k=int(input())
   match k:
      case 1:
        print(client.service.ListOfCountryNamesByName())
      case 2:
        print(client.service.CapitalCity(input()))
      case 3:
        print(client.service.CountryCurrency(input()))
      case 4:
        print(client.service.CountryFlag(input()))
      case 5:
         print(client.service.CountryIntPhoneCode(input()))
      case _:
        print("Error")

def Currency():
  url="http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso"
  client=Client(wsdl=conversion_wsdl(url))
  print(client.service.ListOfCurrenciesByName())

def Languages():
  url="http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso"
  client=Client(wsdl=conversion_wsdl(url))
  print(client.service.ListOfLanguagesByName())

def Book_ISBN_Numbers():
  url="http://webservices.daehosting.com/services/isbnservice.wso"
  client=Client(wsdl=conversion_wsdl(url))
  print("Выберите метод: \n","1) ISBN13 \n", "2) ISBN10")
  k=int(input())
  match k:
      case 1:
        print(client.service.IsValidISBN13(input()))
      case 2:
        print(client.service.IsValidISBN10(input()))
      case _:
        print("Error")

def Temperature():
  url="https://www.w3schools.com/xml/tempconvert.asmx"
  client=Client(wsdl=conversion_wsdl(url))
  print("Выберите метод: \n","1) Из цельсия в фаренгейт \n", "2) Из фаренгейта в цельсий")
  k=int(input())
  match k:
      case 1:
        print(client.service.CelsiusToFahrenheit(input()))
      case 2:
        print(client.service.FahrenheitToCelsius(input()))
      case _:
        print("Error")


def Public_soap_apis():
  while True:
    print("Выберите API: \n","1) Numbers \n", "2) Calculator \n","3) Continents \n","4) Countries\n","5) Currency \n","6) Languages \n","7) Book ISBN Numbers \n","8) Temperature")
    k=int(input())
    match k:
      case 1:
        Numbers()
      case 2:
        Calculator()
      case 3:
        Continents()
      case 4:
        Countries()
      case 5:
        Currency()
      case 6:
        Languages()
      case 7:
        Book_ISBN_Numbers()
      case 8:
        Temperature()
      case _:
        print("Error")



if __name__=='__main__':
  Public_soap_apis()
```
